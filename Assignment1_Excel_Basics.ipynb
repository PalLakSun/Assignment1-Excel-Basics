1.Explain the difference between Relative , Absolute and Mixed Cell Referencing.

Ans-**Relative cell Reference**-Cell reference is automatically changed when copied to other cells.

Use Case-Repeating calculations across rows or columns.

Example-C1=A1+B1 When copied o other cells it's cell reference changes like C2=A2+B2,C3=A3+B3----

**Absolute Cell Reference** -It's cell reference remain fixed no matter where the formula is copied.

Use case-Useful for fixed valueslike tax rates.conversion factors,constants

Example-A1=$A$1
A1*dollarBdollar1-When copied to other cells only A1 changes like A2,A3,--- and dollar($)Bdollar($)1 stays constant.

**Mixed Cell Reference**-It is a combination of relative and absolute cell reference.

Use Case-Helpful in tables or multiplying across both rows and columns

Example-dollarA1-Column 'A' fixed,row changes
Adollar1- Column 'A' changes,Row fixed

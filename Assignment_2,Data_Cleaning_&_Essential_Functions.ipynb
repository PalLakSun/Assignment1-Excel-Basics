 a.What is the difference between TODAY() and NOW() in Excel? Provide an example of when you would use each function.

 b.If cell A1 contains the date 2025-06-10, write a formula using TODAY() to determine how many days are left until that date

 c.Write an Excel formula using NOW() to display the current date and time in the format MM/DD/YYYY HH:MM AM/PM. ́

 d.If a cell contains =TODAY(), what will happen when the worksheet is reopened the next day? Explain

 e.You want to store a static date (today’s date) in a cell without it changing every day. What keyboard shortcut should you use

Ans-

a.Today Function only returns only current date,no time included

ex-18-11-2025

Use TODAY() when you only need the date, such as:

Calculating age

Tracking due dates

Showing today’s date on a report

Counting days remaining until a deadline

Now Function returns current date and time

ex-18-11-2025 10.40 pm

Use NOW() when you need both date and time, such as:

Timestamping entries

Tracking when data was updated

Recording login time or attendance

Real-time dashboards

b. To calculate how many days are left until 2025-06-10 using TODAY(), use this formula:

=A1-Today()

A1=2025-06-10

today()=2025-11-18

result is 2025-06-10 – 2025-11-18 = –161 days

c.=TEXT(NOW(), "mm/dd/yyyy hh:mm")

NOW() → returns current date + time

TEXT() → converts it into your desired format

11/18/2025 22:45
